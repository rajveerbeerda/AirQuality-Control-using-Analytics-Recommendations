```python
import pandas as pd
```

```python
df = pd.concat([pd.read_csv('dataset/cpcb_dly_aq_delhi-2005.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2006.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2007.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2008.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2009.csv')])
```

```python
df.head()
```

```python
df = df.drop(['State', 'City/Town/Village/Area', 'SPM'], axis=1)
df.head()
```

```python
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'])
```

```python
df = df.sort_values(by='Sampling Date')
df.head()
```

```python
locs = ['Nizamuddin_residential', 'Shahadra_industrial', 'Mayapuri_industrial', 
        'N.Y. SCHOOL_residential', 'Siri Fort_residential', 'Pritampura_residential', 
        'Shahzada Bagh_industrial', 'Janakpuri_residential']
```

```python
import csv

for loc in locs:
    data = open(str('city_data 2005-2009/'+loc+'.csv'), 'w')
    rows = []
    header = ['Date', 'NO2', 'SO2', 'RSPM/PM10']
    rows.append(header)
    writer = csv.writer(data)
    
    for i in range(df.shape[0]):
        r1 = df.iloc[i, :]
        r = [str(r1[0].date()), r1[4], r1[3], r1[5]]
        cur_loc = r1[1]
        if cur_loc=='Shahdara':
            cur_loc = 'Shahadra'
            
        if (cur_loc.split())[0].lower()==(loc.split()[0]).split('_')[0].lower():
            rows.append(r)
    writer.writerows(rows)
```
