```python
import pandas as pd
```

```python
df = pd.concat([pd.read_csv('dataset/cpcb_dly_aq_delhi-2011.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2012.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2013.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2014.csv'),
                pd.read_csv('dataset/cpcb_dly_aq_delhi-2015.csv')])
```

```python
df.head()
```

```python
df = df.drop(['Agency', 'City/Town/Village/Area', 'PM 2.5', 'SPM', 'Stn Code', 'State'], axis=1)
df.head()
```

```python
df['Sampling Date'] = pd.to_datetime(df['Sampling Date'])
```

```python
df.head()
```

```python
df = df.sort_values(by='Sampling Date')
df.head()
```

```python
locs = ['Nizamuddin_residential', 'Shahadra_industrial', 'Mayapuri_industrial', 
        'N.Y. SCHOOL_residential', 'Siri Fort_residential', 'Pritampura_residential', 
        'Shahzada Bagh_industrial', 'Janakpuri_residential']
```

```python
import csv

for loc in locs:
    data = open(str('city_data 2011-2015/'+loc+'.csv'), 'w')
    rows = []
    header = ['Date', 'NO2', 'SO2', 'RSPM/PM10']
    rows.append(header)
    writer = csv.writer(data)
    
    for i in range(df.shape[0]):
        r1 = list(df.iloc[i, :])
        r = [str(r1[4].date()), r1[1], r1[3], r1[2]]
        cur_loc = r1[0]

        if (cur_loc.split(',')[0]).split()[0]==(loc.split('_')[0]).split()[0]:
            rows.append(r)
    writer.writerows(rows)
writer.writerows([])
```
